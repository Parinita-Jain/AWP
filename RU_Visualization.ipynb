```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("cars.csv")
df.head(2)
```

```python
df.info()
```

```python
sns.histplot(x=df["price"],label="Histogram Plot",color="green",kde=True)
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.title("Histogram Plot Demo")
plt.legend()
plt.show()

#prices are highly positively skewed.Most probably the outlier is in positive side
#many people are purchasing cars priced 5000-10000
```

```python
sns.boxplot(df["price"],color="orange")

plt.title("BoxPlot")
plt.grid()
plt.show()
```

```python
sns.distplot(df["price"],kde=True)

plt.title("Distribution plot")
plt.grid()
plt.show()
```

```python
sns.countplot(data=df,x="fuel-type")

#people are prefering gas vehichle more over diesel 
```

```python
sns.scatterplot(data=df,x="price",y="engine-size",color="g")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot demo")

plt.show()

#as the size of engine is increasing price is also incrasing 
#hence they are positively correlated
```

```python
sns.scatterplot(data=df,x="price",y="highway-mpg",color="purple")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot demo")

plt.show()
#as the highway mpg is decreasing, price is incrasing 
#hence they are negatively correlated
```

```python
'''A JOINT plot combines a scatter plot with the marginal distributions of the 2 variables it also provides information
about the correlation between the variables '''
```

```python
#JOINT PLOT
sns.jointplot(data=df,x="price",y="highway-mpg",kind="scatter")
```

```python
#KDE PLOT---
'''kde plot visualizes distribution of variable in smooth curve which represents the probability density function'''

sns.kdeplot(data=df,x="price")
```

```python
#PAIR PLOT-----
'''
it shows the relationship between all the pairs of variables in a data sets it includes scatter plot for each pair
and kde plot along the diagonal
'''

sns.pairplot(df[["price","highway-mpg","city-mpg","engine-size"]])
```

```python
#HEAT MAP-----
''' IT Shows the correlation matrix between numerical variables with colors representing the strength of the corrrelation'''

sns.heatmap(df.corr(),annot=True)

#incase of error then first define df
#d=df.select_dtypes(["int","float"])
#sns.heatmap(d.corr(),annot=True,cmap="coolwarm")
```

```python
sns.boxplot(data=df,x="price",y="make")
plt.title("BoxPlot Demo")
plt.grid()
plt.show()

'''A bivarient boxplot is used to compare the distribution of a numerical variable across different categories '''

```
