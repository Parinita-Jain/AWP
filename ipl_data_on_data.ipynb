```python
# Load the provided IPL data CSV file
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
file_path = "data.csv"
ipl_data = pd.read_csv(file_path)

# --- Basic plotting using matplotlib ---
```

```python
# Line Plot: Average runs scored per year
avg_runs_per_year = ipl_data.groupby("year")["runs_scored"].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_runs_per_year.index, avg_runs_per_year.values, color="blue", linestyle="--", marker="o")
plt.title("Average Runs Scored Per Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Runs Scored", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
"""
Line Plot: Average Runs Scored Per Year
Insight:
This chart shows the trend of average runs scored across different IPL seasons.
It provides a high-level view of how scoring patterns have changed over the years. For instance:
If the line rises, it suggests that matches are becoming more high-scoring,
possibly due to improved batting techniques or flatter pitches.
A drop could indicate better bowling strategies or more challenging playing conditions in specific years.

"""
```

```python
# Scatter Plot: Runs scored vs Wickets taken
plt.figure(figsize=(8, 6))
plt.scatter(ipl_data["runs_scored"], ipl_data["wickets_taken"], c="red", alpha=0.6, edgecolor="k")
plt.title("Runs Scored vs Wickets Taken", fontsize=14)
plt.xlabel("Runs Scored", fontsize=12)
plt.ylabel("Wickets Taken", fontsize=12)
plt.grid(alpha=0.3)
plt.show()

"""
Scatter Plot: Runs Scored vs Wickets Taken
Insight:
This plot highlights the relationship between runs scored and wickets taken in individual matches.
Matches with higher runs scored (e.g., above 200) often have fewer wickets taken,
suggesting one or two dominant batting performances.
Matches with lower runs and higher wickets indicate tough bowling conditions or collapses.
Outliers (annotated) show extraordinary performances, such as a match where runs scored exceeded 240.

Why are the points in (near) straight lines?
Discrete Data for Wickets Taken:
Wickets taken in cricket are always discrete integers (0 to 10).
Since there are no fractional wickets, the y-axis (wickets) has fixed levels (e.g., 0, 1, 2, ..., 10).
This makes the data points align horizontally at these levels,
creating the impression of "lines" rather than a continuous spread.
Understanding the Relationship
Expected Relationship:
In cricket, one might expect:
Fewer wickets taken to correlate with higher runs scored (if batters remain at the crease for long periods).
More wickets taken to correlate with lower runs scored (suggesting better bowling performances or batting collapses).
What the Scatter Plot Shows:
No Clear Correlation:
In our scatter plot, runs scored and wickets taken don’t show a strong or consistent relationship.
This indicates that other factors might play a bigger role:
Batting depth: Teams can score high runs even when losing wickets if their middle and lower-order batters contribute.
Match dynamics: A team may deliberately take risks and lose wickets to maximize runs toward the end of an inning.
Bowling quality and conditions: These influence wickets independently of total runs scored.
Analyzing the Data with Correlation
Let’s calculate the actual correlation coefficient (Pearson’s
𝑟
r) between runs_scored and wickets_taken to confirm if there’s any statistical relationship. A value close to 0 would confirm weak or no correlation.
Let me compute this now.



# Calculate the Pearson correlation coefficient between runs_scored and wickets_taken
correlation = ipl_data["runs_scored"].corr(ipl_data["wickets_taken"])
correlation
"""
```

```python
'''
Using FacetGrid in seaborn allows us to create a grid of plots,
splitting the data by a categorical variable. Here’s how we can apply FacetGrid to analyze the ipl_data.csv dataset.

Examples of FacetGrid
1. Runs Scored Distribution Across Teams
Plot the distribution of runs scored (runs_scored) for each team.

python
Copy code
'''
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
ipl_data = pd.read_csv("data.csv")

# FacetGrid for runs scored by team
g = sns.FacetGrid(ipl_data, col="team", col_wrap=4, height=4)
g.map(sns.histplot, "runs_scored", kde=True, color="blue")
g.set_titles("{col_name}")
g.set_axis_labels("Runs Scored", "Frequency")
g.fig.suptitle("Distribution of Runs Scored by Team", y=1.02)
plt.show()
'''
Insights:

Each facet (subplot) shows the distribution of runs scored by a specific team.
You can identify teams with higher scoring ranges or more consistent performance by
analyzing the shape and spread of the histograms.
'''
```

```python
#3D Scatter Plot: Runs Scored, Wickets Taken, and Year

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Load the data
ipl_data = pd.read_csv("data.csv")

# Create a 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with color mapped to teams
scatter = ax.scatter(
    ipl_data["runs_scored"],
    ipl_data["wickets_taken"],
    ipl_data["year"],
    c=ipl_data["year"],
    cmap='viridis',
    s=50,
    alpha=0.8
)

# Adding axis labels and title
ax.set_xlabel("Runs Scored")
ax.set_ylabel("Wickets Taken")
ax.set_zlabel("Year")
ax.set_title("3D Scatter Plot: Runs Scored, Wickets Taken, and Year")

# Adding a color bar to represent years
cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label("Year")

plt.show()
'''
What This Plot Shows
Axis Relationships:

X-axis: Runs scored in a match.
Y-axis: Wickets taken in the same match.
Z-axis: The year of the match.
Year Trend:

Points closer in Z-space (years) might indicate consistency in performance over specific time periods.
You can visually track changes in scoring patterns over the years.
Teams or Match Clusters:

Matches with high runs scored and low wickets taken cluster separately, showing dominant batting performances.
Matches with low runs scored and high wickets taken cluster in different areas, representing bowler-dominant games.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
ipl_data = pd.read_csv("data.csv")

# Calculate the percentage of matches played by each team
team_counts = ipl_data["team"].value_counts()
team_labels = team_counts.index
team_sizes = team_counts.values

# Define colors
colors = ["lightblue", "lightgreen", "orange", "pink", "yellow", "purple", "red", "cyan"]


# Create a pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts=ax.pie(
    team_sizes,
    labels=team_labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops=dict(color="black", fontweight="bold")
)

# Customize the text inside the pie
'''
for text in autotexts:
    text.set_fontsize(12)
    text.set_fontweight("bold")
'''

# Add a legend
ax.legend(wedges, team_labels, title="Teams", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Add a title
ax.set_title("Percentage of Matches Played by Each Team", fontweight="bold", fontsize=16)

plt.show()

'''
you can create a pie chart and then modify it
to a donut chart using matplotlib (Seaborn does not have native support for pie charts).
'''

# Convert Pie Chart into Donut Chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts=ax.pie(
    team_sizes,
    labels=team_labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops=dict(color="black", fontweight="bold"),
    wedgeprops=dict(width=0.4)  # Adjust width for donut chart
)

'''
The ax.pie() function generates a pie chart, with percentages automatically calculated using autopct='%1.1f%%'.

To convert a pie chart into a donut chart,
the wedgeprops=dict(width=0.4) argument is used to reduce the width of the wedges.
'''

# Customize the text inside the donut
'''
for text in autotexts:
    text.set_fontsize(12)
    text.set_fontweight("bold")
'''
# Add a legend
ax.legend(wedges, team_labels, title="Teams", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Add a title
ax.set_title("Percentage of Matches Played by Each Team (Donut Chart)", fontweight="bold", fontsize=16)

plt.show()
```
