```python
'''
The pd.merge() function in pandas is used to combine two DataFrames based on a common column(s) or index.
It’s similar to SQL joins like INNER JOIN, LEFT JOIN, RIGHT JOIN, and OUTER JOIN.

Key Parameters
left and right: The two DataFrames to be merged.
on: The common column(s) to merge on. Must exist in both DataFrames.
left_on and right_on: If the column names in the two DataFrames are different, specify them separately.
how: Type of join. Default is 'inner'. Options are:
'inner': Only matching rows.
'left': All rows from the left DataFrame and matching rows from the right.
'right': All rows from the right DataFrame and matching rows from the left.
'outer': All rows from both DataFrames.
Simple Example: pd.merge()
'''
import pandas as pd

# DataFrame 1: Employees
df1 = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

# DataFrame 1: Employees
df2 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']})

# DataFrame 2: Salaries
df3 = pd.DataFrame({
    'id': [2, 3, 5],
    'salary': [50000, 60000, 45000]
})

print("DataFrame 1:")
print(df1)

print("\nDataFrame 2:")
print(df2)

#Example 1: Using on
#When the column names are the same in both DataFrames:

merged_df = pd.merge(df2, df3, on='id', how='inner')

print("\nMerged DataFrame (using on):")
print(merged_df)
# Merging on the common column 'emp_id' and 'id'
merged_df = pd.merge(df1, df3, left_on='emp_id', right_on='id', how='inner')

print("\nMerged DataFrame:")
print(merged_df)

# Rename 'id' to 'emp_id' for this example
df2.rename(columns={'id': 'emp_id'}, inplace=True)

merged_df = pd.merge(df1, df2, on='emp_id', how='inner')

print("\nMerged DataFrame (with on):")
print(merged_df)

'''
If on or left_on/right_on is not provided, pandas tries to merge on common column names in both DataFrames (if they exist).
'''
```

```python
'''
Using FacetGrid in seaborn allows us to create a grid of plots,
splitting the data by a categorical variable. Here’s how we can apply FacetGrid to analyze the ipl_data.csv dataset.

Examples of FacetGrid
1. Runs Scored Distribution Across Teams
Plot the distribution of runs scored (runs_scored) for each team.

python
Copy code
'''
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
ipl_data = pd.read_csv("data.csv")

# FacetGrid for runs scored by team
g = sns.FacetGrid(ipl_data, col="team", col_wrap=4, height=4)
g.map(sns.histplot, "runs_scored", kde=True, color="blue")
g.set_titles("{col_name}")
g.set_axis_labels("Runs Scored", "Frequency")
g.fig.suptitle("Distribution of Runs Scored by Team", y=1.02)
plt.show()
'''
Insights:

Each facet (subplot) shows the distribution of runs scored by a specific team.
You can identify teams with higher scoring ranges or more consistent performance by
analyzing the shape and spread of the histograms.
'''
```

```python
#3D Scatter Plot: Runs Scored, Wickets Taken, and Year

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Load the data
ipl_data = pd.read_csv("data.csv")

# Create a 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with color mapped to teams
scatter = ax.scatter(
    ipl_data["runs_scored"],
    ipl_data["wickets_taken"],
    ipl_data["year"],
    c=ipl_data["year"],
    cmap='viridis',
    s=50,
    alpha=0.8
)

# Adding axis labels and title
ax.set_xlabel("Runs Scored")
ax.set_ylabel("Wickets Taken")
ax.set_zlabel("Year")
ax.set_title("3D Scatter Plot: Runs Scored, Wickets Taken, and Year")

# Adding a color bar to represent years
cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label("Year")

plt.show()
'''
What This Plot Shows
Axis Relationships:

X-axis: Runs scored in a match.
Y-axis: Wickets taken in the same match.
Z-axis: The year of the match.
Year Trend:

Points closer in Z-space (years) might indicate consistency in performance over specific time periods.
You can visually track changes in scoring patterns over the years.
Teams or Match Clusters:

Matches with high runs scored and low wickets taken cluster separately, showing dominant batting performances.
Matches with low runs scored and high wickets taken cluster in different areas, representing bowler-dominant games.
'''
```

```python
'''
Plot data and a linear regression model fit.
'''
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Creating the dataset
data = {
 'weight': [140, 155, 159, 179, 192, 200, 212],
 'height': [60, 62, 67, 70, 71, 72, 75]
}
df = pd.DataFrame(data)
# Plotting linear regression model
sns.regplot(x='height', y='weight',ci=None, data=df)#,
#line_kws={'color': 'red'})
# Adding labels and title
plt.title("Linear Regression: Weight vs Height")
plt.xlabel("Height (inches)")
plt.ylabel("Weight (lbs)")
# Display the plot
plt.show()
'''
Wider intervals indicate higher uncertainty, while narrower intervals suggest more reliable predictions.
Seaborn computes the confidence intervals using bootstrapping by default.
Bootstrapping involves repeatedly sampling the data (with replacement) and fitting
the regression model to these samples, resulting in a distribution of regression lines.
The confidence interval is derived from this distribution.

'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("cars.csv")
# Define a function to calculate IQR-based statistics for each brand
# Identify outliers using the IQR method
def identify_iqr_outliers(group):
    q1 = group['price'].quantile(0.25)
    q3 = group['price'].quantile(0.75)
    iqr = q3 - q1
    lower_whisker = q1 - 1.5 * iqr
    upper_whisker = q3 + 1.5 * iqr
    outliers = group[(group['price'] < lower_whisker) | (group['price'] > upper_whisker)]
    return outliers[['make', 'price']]

iqr_outliers = df.groupby('make').apply(identify_iqr_outliers).reset_index(drop=True)
iqr_outliers
```

```python
df=pd.read_csv("/content/historical_automobile_sales.csv")
df.head()
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
yearly_sales=df.groupby(df["Date"].dt.year)["Automobile_Sales"].sum()
print(yearly_sales)
```

```python
plt.plot(yearly_sales.index,yearly_sales.values,marker="o",color="b")
plt.xlabel("Year")
plt.ylabel("Automobile Sales")
plt.title("Yearly Automobile Sales")
```

```python
'''
2.PLOT DIFFERENT LINES FOR CATEGORIES OF VEHICLE TYPE
AND ANALYSE THE TREND TO ANSWER THE QUESTION IS THERE
A NOTICEABLE DIFFERENCE IN SALES TRENDS BETWEEN
DIFFERENT VEHICLE TYPES DURING RECESSION PERIODS
'''
yearly_sales_by_type=df.groupby([df["Date"].dt.year,"Vehicle_Type"])["Automobile_Sales"].sum()
print(yearly_sales_by_type)
yearly_sales_by_type=yearly_sales_by_type.unstack(level=1)
print(yearly_sales_by_type)
yearly_sales_by_type=yearly_sales_by_type.fillna(0)
print(yearly_sales_by_type)
for col in yearly_sales_by_type.columns:
    plt.plot(yearly_sales_by_type.index,yearly_sales_by_type[col].values,marker="o",label=col)
plt.xlabel("Year")
plt.ylabel("Automobile Sales")
plt.title("Yearly Automobile Sales by Vehicle Type")
plt.legend()
#
```

```python
'''
To determine if there is a noticeable difference in sales trends between
different vehicle types during recession periods, you can visually inspect
the line chart and focus on the periods that correspond to recession times.
omen indicators to consider while analysing the data:
1. Downturns in Sales: Check if all vehicle types experience a decline
in sales during recession periods or if certain types are more resilient.
2. Recovery Patterns: Analyse how quickly different vehicle types
recover from the recession and if there are variations in the recovery
rates.
3. Relative Performance: Compare the sales trends of different vehicle
types to identify any distinct movements or patterns that differentiate
them during recession periods.
'''
```

```python
'''
3.USE THE MATPLOTLIB LIBRARY TO CREATE A
VISUALIZATION TO COMPARE THE SALES TREND PER
VEHICLE TYPE FOR A RECESSION PERIOD WITH A NONRECESSION PERIOD.

'''
# 1st define recession period as recession period or non recession period by 0 or 1
recession_data=df[df["Recession"]==1]
non_recession_data=df[df["Recession"]==0]
recession_sales_by_type=recession_data.groupby([recession_data["Date"].dt.year,"Vehicle_Type"])["Automobile_Sales"].sum().unstack().fillna(0)
non_recession_sales_by_type=recession_data.groupby([non_recession_data["Date"].dt.year,"Vehicle_Type"])["Automobile_Sales"].sum().unstack().fillna(0)
```

```python
plt.figure(figsize=(12,8))
for col in recession_sales_by_type.columns:
    plt.plot(recession_sales_by_type.index,recession_sales_by_type[col].values,marker="o",label=f"{col} (Recession)")
for col in non_recession_sales_by_type.columns:
    plt.plot(non_recession_sales_by_type.index,non_recession_sales_by_type[col].values,marker="o",label=f"{col} (Non-Recession)")
plt.title("Comparison of Automobile Sales of Vehicle Type by Recession or Non-Recession Types")
plt.xlabel("Year")
plt.ylabel("Automobile Sales")
plt.legend()
plt.show()
```

```python
'''
4.USE THE FUNCTIONALITY OF MATPLOTLIB TO DEVELOP A
SCATTER PLOT TO IDENTIFY THE CORRELATION BETWEEN
AVERAGE VEHICLE PRICE RELATE TO THE SALES VOLUME
DURING RECESSIONS
'''


# Filter data for recession periods
recession_data = df[df['Recession'] == 1]

# Scatter plot: Price vs. Automobile Sales during recessions
plt.figure(figsize=(10, 6))
plt.scatter(recession_data['Price'], recession_data['Automobile_Sales'], alpha=0.7, color='blue')
plt.title('Scatter Plot: Vehicle Price vs. Sales Volume (During Recessions)', fontsize=14)
plt.xlabel('Average Vehicle Price', fontsize=12)
plt.ylabel('Automobile Sales Volume', fontsize=12)
plt.grid(alpha=0.3)
plt.show()

# Calculate correlation
correlation = recession_data['Price'].corr(recession_data['Automobile_Sales'])
correlation
```
