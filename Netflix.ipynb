```python
# Load the provided Netflix data CSV file
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
file_path = "Netflix.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
# missing value in percentage
df.isnull().sum()/len(df)*100
```

```python
# create a dictionary with col names as keys and their datatypes as values
col_data_types = {col: df[col].dtype for col in df.columns}
print(col_data_types)
```

```python
#Find out the number of duplicate rows in the data frame, and drop them. Going forward we will
#use the de-duplicated data frame.
print(df.duplicated().sum())
df=df.drop_duplicates()
print(len(df))
```

```python
#Using De-duplicated data frame from previous assignment question, check if all film titles have
#a unique release date. If not take the oldest date
df['Release Date']=pd.to_datetime(df['Release Date'],errors='coerce')
df=df.sort_values(by='Release Date')
df=df.drop_duplicates(subset='Title',keep='first')
df.head()
```

```python

df["Boxoffice"] = df["Boxoffice"].str.replace("[$,]+", "", regex=True).astype(float)
df.head()

'''
Explanation of changes:

Adding regex=True: The primary issue is with the way the str.replace method is being used.
By default, it operates with literal strings, not regular expressions.
To correctly interpret the [,$]+ pattern as a regular expression targeting
one or more occurrences of '$' or ',', we need to include the regex=True argument.
This allows the replacement to target and remove these special characters within the 'Boxoffice' column's values.
'''
```

```python
#A Box-office hit film means a film which has grossed $20,00,000. Find the number of Box-Office Hit Films.
box_office_hits=df[(df['Series or Movie']=='Movie')&(df['Boxoffice']>=2000000)]
num_box_office_hits=len(box_office_hits)
print(f"Number of Box-Office Hit Films: {num_box_office_hits}")
```

```python
#A critically acclaimed movie means a film that has IMDB rating of atleast 8. Find the number of such films.
critically_acclaimed = df[(df['Series or Movie'] == 'Movie') & (df['IMDb Score'] >= 8)]
num_critically_acclaimed=len(critically_acclaimed)
print(f"Number of Critically Acclaimed Movies: {num_critically_acclaimed}")
```

```python
# Find aLL criticaLLy acclaimed movies which are	Box-off!ce  Hits
critically_acclaimed_hits = df[(df['Series or Movie'] == 'Movie') &
(df['IMDb Score'] >= 8) & (df['Boxoffice']>=2000000)]
print(critically_acclaimed_hits)
```

```python
#Find all critically acclaimed movies which has received at least 5 awards. Order them according to the number of awards.
critically_acclaimed_with_awards = df[(df["Awards Nominated For"]>=5)]
sorted_awards = critically_acclaimed_with_awards.sort_values(by="Awards Nominated For", ascending=False)
print("Critically acclaimed movie with atleast 5 awards")
print(sorted_awards[["Title","IMDb Score","Awards Nominated For"]])
```

```python
#Find all critically acclaimed tv series for United States.
critically_acclaimed_tv_us = df[(df['Series or Movie'] == 'TV Series')
            & (df['IMDb Score'] >= 8)
            & (df["Country Availability"]=="United States")]
print("Critically Acclaimed TV Series from the United States")
print(critically_acclaimed_tv_us[['Title',	"IMDb Score", "Country Availability" ] ])
```

```python
#IMDB Rating is based on IMDB Votes. Find all films which inspite of receiving at least. 100 votes did not receive any award.
df["Awards Nominated For"]=df["Awards Nominated For"].fillna(0)
films_with_votes_no_awards = df[(df['Series or Movie'] == 'Movie') &
         (df['IMDb Votes'] >= 100) & (df["Awards Nominated For"]==0)]
print("Films with at least 100 votes but no awards:")
print(films_with_votes_no_awards[['Title', 'IMDb Votes', 'Awards Nominated For'] ])
```

```python
#A popular film means it is released at Netflix platforms within 1 year.
#Find all Films which were released in Netflix within 2 years of actual release.
#--
# convert release year to date and extract year
df['Release_Year']=pd.to_datetime(df['Release Date'],errors='coerce').dt.year
# convert netflix release datetime to datetime and extract year
df['Netflix Release Year']=pd.to_datetime(df['Netflix Release Date'],errors='coerce').dt.year

YearOffset=(df['Netflix Release Year']-df['Release_Year']).abs()
# filtering logic for movies released within 2 years of actual release
films_released=df[(df['Series or Movie'] == 'Movie') &
         (YearOffset<=2)]
print("Films Released On Netflix Within 2 Years of actual Release")
print(films_released[['Title', 'Release_Year', 'Netflix Release Year']])
```

```python
#Popular Films available in Canada
popular_films_canada=df[df["Country Availability"]=="Canada"]
print("popular Films Available in Canada :")
print(popular_films_canada[['Title', 'Release_Year', 'Netflix Release Year']])
```

```python
# Movies with a runtime of 1 hr
# Convert "Runtime" to numeric by extracting number of minutes
df["Runtime"]=df["Runtime"].str.extract(r'(\d+)').astype(float)
# filter movies with a runtime of atleast 1 hour (60 minutes)
movies_runtime = df[(df["Series or Movie"]=="Movie")&
                    (df["Runtime"]>=1)]
# Display filtered movies--
print("Movies with a runtime of 1 hour or more:")
print(movies_runtime[["Title","Runtime"]])

```

```python
# Series created in more than 1 language
multilingual_tv_series=df[(df["Series or Movie"]=="Series")
                       &(df['Languages'].str.contains(','))]
print("Series created in more than 1 language:")
multilingual_tv_series[["Title","Languages"]]
```

```python
# Movies available in Japan and fall in more than 1 genre
movies_in_japan=df[(df["Series or Movie"]=="Movie")
                   &(df["Country Availability"]=="Japan")
                   &(df['Genre'].str.contains(','))]
print("Movies available in Japan and fall in more than 1 genre:")
movies_in_japan[["Title","Genre"]]
```

```python
# Films with large summary
films_with_large_summary=df[(df["Series or Movie"]=="Movie")&
  (df["Title"].str.len()>50)]
print("Films with large summary:")
print(films_with_large_summary["Title"])
```

```python
# Films with a total rating of atleast 15
films_high_rating = df[(df["Series or Movie"]=="Movie")&
  ((df["IMDb Score"]+df["Hidden Gem Score"])>=15)]
print("Films with a total rating of atleast 15:")
print(films_high_rating[["Title","Hidden Gem Score","IMDb Score"]])
```

```python
# Dramatic Series from Japan
dramatic_series_japan=df[(df["Series or Movie"]=="Series")&
  (df["Genre"].str.contains("Drama"))&
  (df["Country Availability"]=="Japan")]
print("Dramatic Series from Japan:")
dramatic_series_japan[["Title","Country Availability","Genre"]]
```

```python
# films directed by persons whose name starts with "C".
films_directed_by_c=df[(df["Series or Movie"]=="Movie")&
  (df["Director"].str.startswith("C"))]
print("films directed by persons whose name starts with 'C':")
films_directed_by_c[["Title","Director"]]
```

```python
# films directed by Christopher Nolan and ending with 'O'
films_directed_by_christopher_nolan=df[(df["Series or Movie"]=="Movie")&
  (df["Director"]=="Christopher Nolan")&
  (df["Title"].str.endswith("o"))]

print("films directed by Christopher Nolan and ending with 'o':")
films_directed_by_christopher_nolan[["Title","Director"]]
```

```python
# films directed and written by Christopher Nolan
films_directed_and_written_by_christopher_nolan=df[(df["Series or Movie"]=="Movie")&
  (df["Director"]=="Christopher Nolan")&
  (df["Writer"]=="Christopher Nolan")]

print("films directed and written by Christopher Nolan:")
films_directed_and_written_by_christopher_nolan[["Title","Director","Writer"]]
```

```python
# English movie names with IMDb Score more than 8.5
english_movies_high_imdb=df[(df["Series or Movie"]=="Movie")&
  (df["Languages"]=="English")&
  (df["IMDb Score"]>8.5)]

print("English movie names with IMDb Score more than 8.5:")
english_movies_high_imdb[["Title","Languages","IMDb Score"]]
```
